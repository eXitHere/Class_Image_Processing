```python
## open cv = H,W,C
import math 

p3_img = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

print(p3_img.shape)

bit_depth = 4
qlevel = 2**bit_depth
print(qlevel)

# print(p3_img[3])
# 300
for h in range(0,300):
  # 264
  for w in range(0, 264):
    # for c in range(0, 3):
      # print(math.floor(((p3_img[h][w][c]-0)/(255-0))*qlevel))
      # p3_img[h][w][c] = math.floor(((p3_img[h][w][c]-0)/(255-0))*qlevel)
      p3_img[h][w] = math.floor(((p3_img[h][w]-0)/(255-0))*qlevel)

# print(p3_img)
plt.imshow(p3_img, cmap='gray')
```

```python
from matplotlib import cm
from matplotlib.ticker import LinearLocator

p4_img = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

print('Original Dimensions : ', p4_img.shape)

scale_percent = 60
width = int(p4_img.shape[1] * scale_percent / 100)
height = int(p4_img.shape[0] * scale_percent / 100)
dim = (width, height)

p4_img = cv2.resize(p4_img, dim)

print('Resized Dimensions : ', p4_img.shape)

# fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax = plt.axes(projection='3d')

# Make data.
y = p4_img[0:p4_img.shape[0]]
x = p4_img[0:p4_img.shape[1]]
# Plot the surface.
X, Y = np.meshgrid(x, y)

ax.plot_surface( X, Y, p4_img,  cmap='gray')

plt.show()

# plt.imshow(p4_img, cmap='gray')

```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
import numpy as np

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap='gray',
                       linewidth=0, antialiased=False)


plt.show()
```
