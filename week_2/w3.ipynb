```python
cat1 = cv2.imread('cat1.png')
cat2 = cv2.imread('cat2.jpg')

# cat1 = cv2.cvtColor(cat1, cv2.COLOR_BGR2RGB)
# cat2 = cv2.cvtColor(cat2, cv2.COLOR_BGR2RGB)

print(cat1.shape)
print(cat2.shape)


fourcc = cv2.VideoWriter_fourcc(*'MP4V')
# out = cv2.VideoWriter('outpy.avi',cv2.VideoWriter_fourcc('M','J','P','G'), 10, (frame_width,frame_height))
out = cv2.VideoWriter('output.mp4', fourcc, 15.0, (640, 360))

# dst = cv2.addWeighted(cat1, 0.6, cat2, 0.4, 0)
# plt.imshow(dst)

for i in range(40):
  # 0.05 * (i+1), 1 - 0.05 * (i+1)
  # print(0.05 * (i+1), 1 - 0.05 * (i+1))
  dst = cv2.addWeighted(cat1, 0.025 * (i+1), cat2, 1 - 0.025 * (i+1), 0)
  # plt.imshow(dst)
  out.write(dst)

for i in range(40):
  # 0.05 * (i+1), 1 - 0.05 * (i+1)
  # print(0.05 * (i+1), 1 - 0.05 * (i+1))
  dst = cv2.addWeighted(cat1, 1 - 0.025 * (i+1), cat2, 0.025 * (i+1), 0)
  # plt.imshow(dst)
  out.write(dst) 

out.release()

# dst = cv2.addWeighted(cat1, 0.5, cat2, 0.5, 0)

# plt.imshow(dst)
```
